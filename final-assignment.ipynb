## Load images



```python
imgL = cv.imread('img/allee_l.png', 1)
imgR = cv.imread('img/allee_r.png', 1)
print(imgL[340])

imgL = cv.cvtColor(imgL, cv.COLOR_BGR2RGB)
imgR = cv.cvtColor(imgR, cv.COLOR_BGR2GRAY)
```
